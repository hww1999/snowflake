```python
import matplotlib.pyplot as plt
import numpy as np

# Data
warehouse_sizes = ['X-Small', 'Small', 'Medium', 'Large', 'X-Large']
modeling_data = {
    'Hour': [55.333333, 56.666667, 53.333333, 50.0, 56.666667],
    'Day': [30.333333, 31.0, 32.0, 31.0, 31.0],
    'Week': [13.666667, 13.333333, 9.8, 7.933333, 10.066667],
    'Month': [16.333333, 8.033333, 9.8, 7.933333, 10.066667]
}
prediction_data = {
    'Hour': [600.333333, 619.333333, 638.0, 643.0, 637.333333],
    'Day': [29.666667, 31.0, 32.0, 31.0, 31.0],
    'Week': [11.0, 8.6, 10.666667, 8.033333, 7.6],
    'Month': [13.333333, 8.633333, 10.666667, 8.033333, 7.6]
}

timeframes = ['Hour', 'Day', 'Week', 'Month']
colors = ['#0065A2', '#0065A2', '#0065A2', '#0065A2']  # Same color for modeling across plots
prediction_colors = ['#00B0BA', '#00B0BA', '#00B0BA', '#00B0BA']  # Same color for prediction across plots

fig, axs = plt.subplots(1, 4, figsize=(16, 4))  # One row with four subplots
for i, timeframe in enumerate(timeframes):
    ax = axs[i]
    ax.bar(warehouse_sizes, modeling_data[timeframe], label='Model Training', color=colors[i])
    ax.bar(warehouse_sizes, prediction_data[timeframe], bottom=modeling_data[timeframe], label='Prediction',
           color=prediction_colors[i])

    ax.set_xlabel('Warehouse Size')
    ax.set_ylabel('Time (s)')
    ax.set_title(f'{timeframe} Data')
    
    if timeframe != 'Hour':
        ax.set_ylim([0, 70])  # Set y-axis limits from 0 to 60 for Day, Week, Month plots
    else:
        ax.set_ylim([0, 750])  # Set consistent y-axis limits for the Hour plot
    
    if i != 3:
        ax.legend().set_visible(False)  # Hide legends for Hour, Day, and Week plots

# Show legend only in the last plot (Month)
axs[-1].legend()

plt.tight_layout()
plt.show()
```
