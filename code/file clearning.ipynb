```python
import json
```

```python
f_2010 = open('cli_2010.json')
f_2013 = open('cli_2013.json')
f_2016 = open('cli_2016.json')
f_2019 = open('cli_2019.json')
f_2020 = open('cli_2020.json')

rs = {"data":[]}
# returns JSON object as 
# a dictionary
data1 = json.load(f_2010)
data2 = json.load(f_2013)
data3 = json.load(f_2016)
data4 = json.load(f_2019)
data5 = json.load(f_2020)
rs["data"] = data1['data'] + data2['data'] + data3['data'] + data4['data'] + data5['data']


with open("cli_total.json", "w") as outfile:
    outfile.write(json.dumps(rs,indent=4))

# Closing file
f_2010.close()
f_2010.close()
f_2010.close()
f_2010.close()
f_2010.close()
```

```python
import pandas as pd

df = pd.read_json('cli_total.json')
df = pd.json_normalize(df['data'])
```
